## Asynchronous

First, with addition of *p5* pattern.

```python
old_network_5_async = hopfield.Hopfield(25, (5, 5), synchronous=False)
old_network_5_async.train(train[:5])
```

```python
for i in range(5):
  plt.imshow(old_network_5_async.update(train[i]), cmap='plasma')
  plt.show()
```
