Using the Bayes Theorem find posterior distribution for the $\mu$ paramter.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd
```

Posterior expectation value and variance accroding to the derivations.

```python
def post_var(N, sigma2, sigma2_0):
  return (N/sigma2+1/sigma2_0)**-1
```

```python
def post_expectation(N, sigma2, sigma2_0, X_sum, mu_0):
  post_sigma2 = post_var(N, sigma2, sigma2_0)
  return post_sigma2 * (X_sum/sigma2 + mu_0/sigma2_0)
```

Plotting a single distribution for given $N$ and set of $X$s.

```python
def plot_post(N, X, sigma2, sigma2_0, mu_0, plot_lims):
  X_plot = np.arange(plot_lims[0],plot_lims[1],0.01)
  ps2 = post_var(N, sigma2, sigma2_0)
  pmu = post_expectation(N, sigma2, sigma2_0, X.sum(), mu_0)
  plt.plot(X_plot, norm.pdf(X_plot, loc=pmu, scale=ps2), label=f'N = {N}')
```

Plotting plots for all the $N$s.

```python
def plot_post_dists(Ns=[5, 10, 20, 50], sigma2=2, mu_X=5, sigma2_0=10, mu_0=0, plot_lims=(-10, 20)):
  for n in Ns:
    X = np.random.normal(size=(n,), loc=mu_X, scale=sigma2)
    plot_post(n, X, sigma2, sigma2_0, mu_0, plot_lims=plot_lims)
  plt.plot([mu_X, mu_X], [0,10], label=f'True value')
  plt.plot(np.arange(plot_lims[0],plot_lims[1],0.01), norm.pdf(np.arange(plot_lims[0],plot_lims[1],0.01), loc=mu_0, scale=sigma2_0), label=f'Prior')
  plt.title(r'Posterior distributions for $\mu$')
  plt.xlabel(r'$\mu$')
  plt.ylabel('Posterior density')
  plt.legend()
  plt.show()
```

Prior with $\mu_0=0$ and $\sigma_0^2=10$.

```python
plot_post_dists(plot_lims=(0,7))
```

Prior with $\mu_0=-2$ and $\sigma_0^2=0.5$.

```python
plot_post_dists(mu_0=-2, sigma2_0=0.5, plot_lims=(-5,6))
```

Prior with $\mu_0=10$ and $\sigma_0^2=0.5$.

```python
plot_post_dists(mu_0=10, sigma2_0=0.5, plot_lims=(4,12))
```

Conluding, despite different priors, the posterior distributions converges with growing $N$ to the true value.
